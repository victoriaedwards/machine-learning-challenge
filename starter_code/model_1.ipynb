```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.columns
```

```python
df=df.drop(['koi_period_err1', 'koi_time0bk_err1', 'koi_impact_err2','koi_duration_err1','koi_depth_err1','koi_prad_err2','koi_insol_err2','koi_steff_err2','koi_srad_err2','dec'], axis=1)
```

```python
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
#selected_features = df[['koi_pdisposition', 'koi_score', 'koi fpﬂag nt', 'koi fpﬂag ss', 'koi depth']]
X=df.drop('koi_disposition', axis=1)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y=df['koi_disposition']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y,test_size=0.2, random_state=43 )
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
#from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

# Train the Model



```python
#Train the model
from sklearn.svm import SVC 
svc_model = SVC(kernel='linear')
svc_model.fit(X_train, y_train)
pred_svc = svc_model.predict(X_test)
```

```python
#pred_rfc[:20]
```

```python
print(f"Training Data Score: {svc_model.score(X_train, y_train)}")
print(f"Testing Data Score: {svc_model.score(X_test, y_test)}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, pred_svc))
print(confusion_matrix(y_test, pred_svc))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = [{'C': [1, 5,9, 10,11, 15, 20], 'kernel':['linear'],
              'gamma': [0.00001, 0.0001, 0.001, 0.01]}, {'C': [1, 5, 9,10,11, 15, 20], 'kernel': ['rbf'],
              'gamma': [0.00001, 0.0001, 0.001, 0.01]}]
grid = GridSearchCV(svc_model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
svc_model2 = SVC(C=20, kernel='linear', gamma=1e-05)
svc_model2.fit(X_train, y_train)
pred_svc2 = svc_model2.predict(X_test)
```

```python
print(classification_report(y_test, pred_svc2))
print(confusion_matrix(y_test, pred_svc2))
```

```python
print(f"Training Data Score: {svc_model2.score(X_train, y_train)}")
print(f"Testing Data Score: {svc_model2.score(X_test, y_test)}")
```
